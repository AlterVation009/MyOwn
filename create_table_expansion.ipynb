```python
import pandas as pd

df_pandas = pd.read_excel(f"{notebookutils.nbResPath}/builtin/Files/PIPELINE.xlsx", skiprows=5)

df_spark = spark.createDataFrame(df_pandas)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, StringType, DateType

df_spark1 = df_spark.withColumnRenamed("Num Región", "Num_Region") \
                   .withColumnRenamed("Objetivo Aperturas", "Objetivo_Aperturas")\
                   .withColumnRenamed("Aperturas ", "Aperturas")



# Castear las columnas
df_final = df_spark1.withColumn("Num_Region", col("Num_Region").cast(IntegerType())) \
                   .withColumn("DO", col("DO").cast(StringType())) \
                   .withColumn("Región", col("Región").cast(StringType())) \
                   .withColumn("Fecha", col("Fecha").cast(DateType())) \
                   .withColumn("Año", col("Año").cast(IntegerType())) \
                   .withColumn("Objetivo_Aperturas", col("Objetivo_Aperturas").cast(IntegerType())) \
                   .withColumn("Aperturas", col("Aperturas").cast(IntegerType()))

# Verificar esquema actualizado
df_final.printSchema()
# df_spark.show()
df_final.write.format("delta").mode("overwrite").saveAsTable('LZ_LH_Main.kpis_expansion')
```
