### Gets the rolling 70 days up until the day we cut over to sybase 17 for FMovarti.

```python
# With Spark SQL, Please run the query onto the lakehouse which is from the same workspace as the current default lakehouse.
df = spark.sql("SELECT * FROM LZ_LH_Main.fmovarti_hist_old WHERE ATMFECHA >= DATEADD(DAY, -70, GETDATE()) and ATMFECHA <= '2025-01-27' ")
df.write.mode('overwrite').parquet("Files/User Created Data/fmovarti_hist_api")
```

```python
%%sql
CREATE TABLE starting_inventory_records AS 
WITH first_dates AS (
    SELECT 
        TCLAVE,
        ICLAVE,
        MIN(ATMFECHA) AS first_date
    FROM finvhist_calculated_bkp_v2
    GROUP BY TCLAVE, ICLAVE
)
SELECT 
    f.TCLAVE,
    f.ICLAVE,
    f.ATMFECHA,
    f.movements_sum,
    f.movements_sum_costo,
    cantidad,
    end_of_day,
    balance,
    costo_unitario,
    inventario_inicial_costo,
    inventario_final_costo
FROM finvhist_calculated_bkp_v2 f
JOIN first_dates fd
  ON f.TCLAVE = fd.TCLAVE
  AND f.ICLAVE = fd.ICLAVE
  AND f.ATMFECHA = fd.first_date;
```
